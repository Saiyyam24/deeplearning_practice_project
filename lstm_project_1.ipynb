```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([text])
```

```python
tokenizer.word_index
```

```python
input_sequences = []
for sentences in text.split('\n'):
 tokenize_sentence = tokenizer.texts_to_sequences([sentences])[0]
 for i in range(1,len(tokenize_sentence)):
  input_sequences.append(tokenize_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
padded_input_sequences = pad_sequences(input_sequences,maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
```

```python
X
```

```python
y = padded_input_sequences[:,-1]
```

```python
y
```

```python
X.shape
```

```python
y.shape
```

```python
len(tokenizer.word_index)
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes = 200)
```

```python
y
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense,Embedding
```

```python
model = Sequential()
```

```python
model.add(Embedding(200,100, input_length=41))
model.add(LSTM(150))
model.add(Dense(200,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=100)
```

```python
message = "Python"


# tokenize
tokenize_text = tokenizer.texts_to_sequences([message])[0]
# Padding
padded_text = pad_sequences([tokenize_text],maxlen = 42,padding='pre')

```

```python
# Predict
import numpy as np
pos = np.argmax(model.predict(padded_text))
```

```python
for word,index in tokenizer.word_index.items():
  if index == pos:
    print(word)
```
